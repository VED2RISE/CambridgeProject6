## Just for the sake of any future use, we should impute the missing data.
Specially for that I found datasets that contain cities with the provinces where they are located.
But first let's examine the data. It is worth creating a dictionary with all the cities and provinces to which they belong. Perhaps there are lines in which there is a city x and a province y, but there are also lines in which there is only city x, and the province is listed as a blank.

```python
city_province = {}
for index, row in clean.iterrows():
  city = row["City"]
  province = row["State_Province"]
  if pd.notna(province):
    city_province[city] = province
```

```python
def fill_province(row):
  if pd.isna(row["State_Province"]):
    return city_province.get(row["City"], row["State_Province"])
  return row["State_Province"]

clean["State_Province"] = clean.apply(fill_province, axis=1)
```

The function below imputes the missing provinces from the datasets

```python
def formatter(source, clean, is_file=True):
    if is_file:
        df = pd.read_csv(source)
    else:
        df = source

    df = df.filter(items=["city", "admin_name"])
    df.rename(columns={'city': 'City', 'admin_name': 'State_Province'}, inplace=True)


    province_dict = dict(zip(df['City'], df['State_Province']))


    def fill_provinces(row):
        if pd.isna(row['State_Province']):
            return province_dict.get(row['City'], row['State_Province'])
        return row['State_Province']

    clean['State_Province'] = clean.apply(fill_provinces, axis=1)


formatter("nl.csv", clean)
formatter("worldcities.csv", clean)
```

```python
def currency_to_float(currency_str):
    if isinstance(currency_str, float):
        return currency_str

    currency_str = currency_str.replace('$', '').replace(',', '').strip()

    if currency_str.startswith('(') and currency_str.endswith(')'):
        currency_str = '-' + currency_str[1:-1]

    return float(currency_str)

cleaned['Total Revenue'] = cleaned['Total Revenue'].apply(currency_to_float)
```

## Now we need to perform clustering on the selected features.
To do this, we'll create a pipeline, calculate the silhouette score, and the elbow score.

```python
required_columns = ['Age', 'Recency', 'Frequency', 'CLV', 'Average_Unit_Cost']
if not all(column in cleaned.columns for column in required_columns):
    raise ValueError(f"DataFrame is missing one of the required columns: {required_columns}")

if cleaned[required_columns].isnull().any().any():
    raise ValueError("DataFrame contains missing values. Please handle them before clustering.")

X = cleaned[required_columns]

if isinstance(X, pd.Series):
    X = X.to_frame()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sse = []
k_range = range(1, 6)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, algorithm='elkan')
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
silhouette_avg = []
for k in range(2, 6):
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('kmeans', KMeans(n_clusters=k, random_state=42))
    ])
    cluster_labels = pipeline.fit_predict(X)
    silhouette_avg.append(silhouette_score(X, cluster_labels))

plt.figure(figsize=(10, 6))
plt.plot(range(2, 6), silhouette_avg, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Analysis for Optimal k')
plt.grid(True)
plt.show()
```

## As a result of extensive parameter tuning, I decided to settle on this number of clusters and these features (3).
This selection strikes a compromise between the two metrics and provides clear visualization for further analysis.

```python
optimal_k = 3
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(n_clusters=optimal_k, random_state=42))
])

cleaned['Cluster'] = pipeline.fit_predict(X)
cleaned_copy = cleaned.copy()
cleaned_copy = cleaned_copy.drop("Customer ID", axis=1)

pca = PCA(n_components=2) #VISUALISING THE OUTPUT
pca_result = pca.fit_transform(StandardScaler().fit_transform(cleaned_copy))

plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    cluster_data = pca_result[cleaned['Cluster'] == cluster]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster}')

plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clustering with PCA for Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

Visualizing the same data with t-SNE. We can sharply observe 3 defined clustres.

```python
tsne = TSNE(n_components=2, random_state=42)
cleaned_copy_2 = cleaned.copy()
tsne_result = tsne.fit_transform(StandardScaler().fit_transform(cleaned_copy_2))

plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    cluster_data = tsne_result[cleaned['Cluster'] == cluster]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster}')

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('K-Means Clustering with t-SNE for Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

```python
cluster_labels = pipeline.named_steps['kmeans'].labels_
clustered_customers = pd.DataFrame({
    'Customer ID': cleaned['Customer ID'],
    'cluster': cluster_labels
})

clustered_customers = clustered_customers.merge(cleaned[['Customer ID', "CLV", "Average_Unit_Cost", "Frequency", "Recency", "Age"]], left_on='Customer ID', right_on='Customer ID')


plt.figure(figsize=(10, 6))
sns.boxplot(x='cluster', y='CLV', data=clustered_customers)
plt.xlabel('Cluster')
plt.ylabel('Total Revenue')
plt.title('Boxplot of Total Revenue by Cluster')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='cluster', y='Age', data=clustered_customers)
plt.xlabel('Cluster')
plt.ylabel('Age')
plt.title('Boxplot of Age by Cluster')
plt.show()

sns.boxplot(x='cluster', y='Frequency', data=clustered_customers)
plt.xlabel('Cluster')
plt.ylabel('TFrequency')
plt.title('Boxplot of Frequency by Cluster')
plt.show()

sns.boxplot(x='cluster', y='Recency', data=clustered_customers)
plt.xlabel('Cluster')
plt.ylabel('Recency')
plt.title('Boxplot of Recency by Cluster')
plt.show()

sns.boxplot(x='cluster', y='Average_Unit_Cost', data=clustered_customers)
plt.xlabel('Cluster')
plt.ylabel('Avergae Unit Cost')
plt.title('Boxplot of Average Unit COost by Cluster')
plt.show()
```
